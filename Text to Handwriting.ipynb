```python
import pywhatkit as kit
from PIL import Image

#Taking input from user
Handwritten = input("Enter your text to convert in Handwriting :")

#converting to handwritten to saving
kit.text_to_handwriting(Handwritten, save_to = "mx6coding.png")

#To open the image
Image.open("mx6coding.png")
```
